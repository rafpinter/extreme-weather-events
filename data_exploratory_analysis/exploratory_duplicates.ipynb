# Data Exploratory Analysis

## Duplicated rows

The objective of this notebook is try to find if there are rows with the same features that are duplicated in different labels.

```python
import pandas as pd

pd.set_option('display.max_columns', None)
```

```python
label_col = "Label"
```

```python
raw_data = pd.read_csv("../../data/train.csv")
```

```python
def normalize_data(data):
    mu = data[:, :].mean(axis=0)
    sigma = data[:, :].std(axis=0)
    data[:, :] = (data[:, :] - mu) / sigma
    return data
```

Trying to find duplicated rows

```python
raw_data[raw_data.columns[~raw_data.columns.isin(["time", "SNo"])]].columns
```

```python
raw_data_count = raw_data.groupby(['lat', 'lon', 'TMQ', 'U850', 'V850', 'UBOT', 'VBOT', 'QREFHT', 'PS',
       'PSL', 'T200', 'T500', 'PRECT', 'TS', 'TREFHT', 'Z1000', 'Z200', 'ZBOT']).Label.nunique().reset_index(name="Label_count")
```

```python
raw_data_count
```

```python
raw_data_count = pd.merge(raw_data, raw_data_count, how='left', on=['lat', 'lon', 'TMQ', 'U850', 'V850', 'UBOT', 'VBOT', 'QREFHT', 'PS',
       'PSL', 'T200', 'T500', 'PRECT', 'TS', 'TREFHT', 'Z1000', 'Z200', 'ZBOT'])
```

```python
raw_data_count.groupby(["Label", "Label_count"]).size().reset_index(name="count_duplicates")
```

There are multiple rows with the same values for all features that belong to different classes
